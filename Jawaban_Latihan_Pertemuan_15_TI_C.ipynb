```python
#1. Terdapat list atau array dengan elemen : 100,121,50,14,16,20. Buatlah list tersebut dengan
#menggunakan numpy dan buatlah grafik point atau scatter dengan data elemen-elemen pada list
#tersebut.

import matplotlib
import matplotlib.pyplot as plt
import numpy as np
x=np.array([100,121,50,14,16,20])
plt.plot(x,"ob")
plt.show()
```

```python
#2. Terdapat dictionary dengan data berupa nilai Matematika 10 siswa, yaitu :
#Burhanuddin : 50
#Anwar : 80
#Amin : 80
#Siti : 70
#Aminah : 100
#Nia : 88
#Buatlah grafik garis dengan data tersebut dengan diberi judul, label sumbu X dan label sumbu Y.
#Judul grafik : “Nilai Matematika”, Label sumbu X : “Nama“ dan Label sumbu Y : “Nilai”.
            
import matplotlib 
import matplotlib.pyplot as plt
import numpy as np
dict={"Burhanuddin":50,"Anwar":80,"Amin":80,"Siti":70,"Aminah":100,"Nia":88}
names=list(dict.keys())
values=list(dict.values())
plt.plot(names,values)
plt.suptitle("Nilai Matematika")
plt.xlabel("Nama")
plt.ylabel("Nilai")
plt.show()
```
