```python
#1. Terdapat list atau array dengan elemen : 10,11,90,4,6,10. Buatlah list tersebut dengan
#menggunakan numpy dan buatlah grafik garis dengan data elemen-elemen pada list tersebut.

import numpy as np
import matplotlib 
import matplotlib.pyplot as pyt
x=np.array([10,11,90,4,6,10])
pyt.plot(x)
pyt.show()
```

```python
#2. Terdapat dictionary dengan data berupa jumlah warga di RW 02 pada setiap RT nya:
#RT.02 : 250
#RT.03 : 200
#RT.04 : 210
#RT.05 : 220
#RT.06 : 300
#Buatlah grafik batang data tersebut dengan diberi judul, label sumbu X dan label sumbu Y.
#Judul grafik : “Jumlah warga di RW 02”, Label sumbu X : “RT“ dan Label sumbu Y : “Jumlah
#Penduduk”.

import matplotlib 
import matplotlib.pyplot as pyt
rt={"rt.02":250,"rt.03":200,"rt.04":210,"rt.05":220,"rt.06":300}
names=list(rt.keys())
values=list(rt.values())
pyt.bar(names,values)
pyt.suptitle("Jumlah warga di RW 02")
pyt.xlabel("RT")
pyt.ylabel("Jumlah Penduduk")
```
